#### extracting text from the audio file

```python
import speech_recognition as sr
from googletrans import Translator

translator = Translator()

def choose_lang():
    indian_languages = {
    'hi-IN': 'Hindi',
    'bn-IN': 'Bengali',
    'te-IN': 'Telugu',
    'mr-IN': 'Marathi',
    'ta-IN': 'Tamil',
    'gu-IN': 'Gujarati',
    'kn-IN': 'Kannada',
    'ml-IN': 'Malayalam',
    'pa-IN': 'Punjabi',
    'ur-IN': 'Urdu',
    'or-IN': 'Odia',
    'as-IN': 'Assamese'
}

    print("Select the desired language by entering the corresponding number:")
    for i, lang_code in enumerate(indian_languages.items(), 1):
        print(f"{i}. {lang_code[1]} ({lang_code[0]})")

    # Ask the user to input a number for the desired language
    language_choice = input("Enter the number corresponding to the desired language: ")

    # Validate user input and choose the language
    if language_choice.isdigit() and 1 <= int(language_choice) <= len(indian_languages):
        chosen_language = list(indian_languages.keys())[int(language_choice) - 1]
    else:
        print("\nPress Enter to continue with English...")
        chosen_language = 'en-IN'

    return chosen_language


def save_audio(filename="user_input.wav"):
    recognizer = sr.Recognizer()
    recognizer.pause_threshold = 0.8

    with sr.Microphone() as source:
        print("Please speak something...")
        recognizer.adjust_for_ambient_noise(source)  
        audio = recognizer.listen(source)


    with open(filename, "wb") as f:
        f.write(audio.get_wav_data())

    return filename

def process_audio(audio_file, chosen_lang):
    recognizer = sr.Recognizer()
    recognizer.pause_threshold = 0.8

    # Process the audio file
    with sr.AudioFile(audio_file) as source:
        audio = recognizer.record(source)

    try:
        # chosen_language = 'te-IN'
        text = recognizer.recognize_google(audio, language=chosen_lang)
        # detected_language = translator.detect(text).lang
        if chosen_lang!='en-IN':
            translated_text = translator.translate(text, dest='en').text
            return {
                'recognized_text': text,
                'translated_text': translated_text
            }
        else:
            return {"Text": text
            }
    except sr.UnknownValueError:
        text = "Sorry, I couldn't understand what you said."
        return text
    except sr.RequestError as e: 
        text = "Sorry, I couldn't request results from the Google Speech Recognition service; {0}".format(e)
        return text
if __name__ == '__main__':
    chosen_lang=choose_lang()
    audio_file = save_audio()
    result = process_audio(audio_file, chosen_lang)
    print(result)
```
